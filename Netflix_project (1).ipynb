## **Pivot** **Table**

```python
pivot_type = pd.pivot_table(
    NF,
    index="type",
    values="title",
    aggfunc="count")
print(pivot_type)
```

```python
pivot_country = pd.pivot_table(
    NF,
    index="country",
    values="title",
    aggfunc="count").sort_values(by="title", ascending=False).head(10)
print(pivot_country)
```

```python
pivot_year_type = pd.pivot_table(
    NF,
    index="release_year",
    columns="type",
    values="title",
    aggfunc="count",
    fill_value=0)
print(pivot_year_type)
```
